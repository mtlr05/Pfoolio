```python
import pandas as pd
import numpy as np
import re
```

```python
def conv(value,currency,rate):
    if currency == 'USD':
        return value*rate
    else:
        return value


def cash(file):
    word = 'Total(in CAD)'
    with open(file, 'r') as fp:
        # read all lines in a list
        lines = fp.readlines()
        for line in lines:
            # check if string present on a current line
            if line.find(word) != -1:
                #print(word, 'string exists in file')
                #print('Line Number:', lines.index(line))
                return float(re.findall("\d+\.\d+", line)[1])
```

```python
file1='MyHoldings_22964041_15-Mar-2023.csv'
file2='MyHoldings_22135047_15-Mar-2023.csv'
usd2cad=1.3663

#cash1 = pd.DataFrame.from_dict(dict)
Cash = pd.DataFrame.from_dict({'Asset class':['Cash','Cash'],'Symbol':['CASH','CASH'],'Market value':[cash(file1),cash(file2)]})
#cash1=cash(file1)
#cash2=cash(file2)

print(cash)

columns=['Asset class','PROD Type/IND. Sector','Description','Symbol','Quantity','Average cost','Currency','Total cost','y','Current price','z','Market value','a','Unrealized gain/loss','b' ,'Unrealized gain/loss %','Portfolio%','FX rate','Previous close price','c' ,'Maturity date','Days to maturity','Annualized income', 'd','Yield','Indicated annual dividend', 'x','Dividend frequency','Ex-dividend date','Price type','Cost type']
usecols=['Asset class','PROD Type/IND. Sector','Description','Symbol','Quantity','Average cost','Total cost','Current price','Market value','Unrealized gain/loss','Unrealized gain/loss %','Portfolio%','FX rate','Previous close price','Maturity date','Days to maturity','Annualized income','Yield','Indicated annual dividend', 'Currency','Dividend frequency','Ex-dividend date','Price type','Cost type']
df1 = pd.read_csv(file1,header=9,names=columns,index_col=False,usecols=usecols)
df2 = pd.read_csv(file2,header=9,names=columns,index_col=False,usecols=usecols)
index1 = (df1["Asset class"]=="Equities") | (df1["Asset class"]=="Options")
index2 = (df2["Asset class"]=="Equities") | (df2["Asset class"]=="Options")
TFSA_22964041 = df1.loc[index1]
RRSP_22135047 = df2.loc[index2]

PFolio = pd.concat([TFSA_22964041, RRSP_22135047,Cash], ignore_index=True, sort=False)
#print(PFolio)
#PFolio['Market value CAD'] = PFolio.apply(lambda x: conv(x['Market value'], x['Currency'],usd2cad), axis=1)
total =  PFolio['Market value'].sum()
PFolio['Percentage'] = PFolio.apply(lambda x: round(x['Market value']/total*100,1), axis=1)

PF = PFolio.groupby('Symbol')['Market value','Percentage'].sum()

print(PF.sort_values(by='Percentage',ascending=False))

#print(PFolio.groupby('Symbol').sum())
```
